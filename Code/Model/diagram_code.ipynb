```python
from diagrams.aws.iot import IotSensor, IotCore, IotMqtt, IotAnalyticsChannel, IotAnalyticsDataSet, IotAnalyticsDataStore,IotAnalyticsPipeline
from diagrams import Cluster, Diagram
```

```python
with Diagram('Diagrama de fluxo de dados', show=False) as diag:
  with Cluster("Edge"):
    sensor = IotSensor("sensor")
    mqtt = IotMqtt("computer/temperature")
  with Cluster("Cloud"):
    iotCore = IotCore("IoTCore")
    with Cluster("AWS analytics"):

      channel = IotAnalyticsChannel("Channel")
      pipeline = IotAnalyticsPipeline("Pipeline")
      datastore = IotAnalyticsDataStore("Datastore")
      dataset = IotAnalyticsDataStore("Dataset")

  sensor >> mqtt >> iotCore >> channel >> pipeline >> datastore >> dataset
diag
```

```python
from diagrams.programming.flowchart import Document, Database, StartEnd, Preparation, Sort,StoredData  
```

```python
with Diagram('Diagrama de obtenção de dados', show=False) as diag:
  with Cluster("Edge"):
    documento = Document("Site")
    database = IotAnalyticsDataStore("CSV")
    inicio = StartEnd("Inicio")
    ordem = Sort("Organização")
    tratamento = Preparation("Tratamento")
    dados_tratados = IotAnalyticsDataStore("Dados tratados")
    fim = StartEnd("Fim")

  documento >> database >> inicio >> ordem >> tratamento >> dados_tratados >> fim
diag
```
